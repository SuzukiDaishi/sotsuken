```python
%matplotlib inline
import matplotlib.pyplot as plt
import IPython.display as ipd
import wave
import util
import numpy as np
from keras.models import load_model
from keras_contrib.layers.normalization.instancenormalization import InstanceNormalization
from keras_contrib.layers.convolutional.subpixelupscaling import SubPixelUpscaling
```

```python
_, coded_sps_mean_A, coded_sps_std_A, coded_sps_max_A, log_f0s_mean_A, log_f0s_std_A = util.loadPickle('./cache36_suzuki.pkl')
_, coded_sps_mean_B, coded_sps_std_B, coded_sps_max_B, log_f0s_mean_B, log_f0s_std_B = util.loadPickle('./cache36_midoma.pkl')
src_path = 'コーヒーメーカー.wav'
model_path = 'predict/log_cyclegan_AtoB_50.h5'
custom_objects = {
    'InstanceNormalization': InstanceNormalization,
    'SubPixelUpscaling': SubPixelUpscaling
}
LEN = 5
```

```python
wave = np.pad(util.loadWave(src_path), (60*100, 60*100), 'constant')
pwav = util.wavePadding(wave)
f0, sp, ap = util.worldDecompose(pwav)
psf0 = util.pitchConversion(f0, log_f0s_mean_A, log_f0s_std_A, log_f0s_mean_B, log_f0s_std_B)
g_AB = load_model(model_path, custom_objects, compile=False)
```

```python
coded_sp = util.worldEncodeSpectralEnvelop(sp)
coded_sp_t = coded_sp.T
coded_sp_norm = (coded_sp_t - coded_sps_mean_A) / coded_sps_max_A
coded_sp_norm = coded_sp_norm[:, :128*LEN]
coded_sp_norm = coded_sp_norm.reshape(LEN, 36, 128)
dist = g_AB.predict(coded_sp_norm)
dist = dist.reshape(36, 128*LEN)
```

```python
dist = dist * coded_sps_max_B + coded_sps_mean_B
dist = dist.T
dist = np.ascontiguousarray(dist)
dist = util.worldDecodeSpectralEnvelop(dist)
dist = util.worldSpeechSynthesis(psf0[:dist.shape[0]], dist[:dist.shape[0]], ap[:dist.shape[0]])
dist = np.where(dist >  1.,   1., dist)
dist = np.where(dist < -1., -1., dist)
```

```python
plt.plot(pwav[:dist.shape[0]])
plt.plot(dist)
```
